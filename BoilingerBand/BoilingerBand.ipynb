## BoilingerBand

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
symbol=["RELIANCE.NS"]
start_date='1995-11-29'
end_date=datetime.now()
```

```python
reliance=yf.download(symbol , start_date , end_date, multi_level_index=False)
```

```python
def BoilingerBand(data , window=n):
    data['Moving Average']=data.Close.rolling(window=n).mean()
    data['Standard Deviation']=data.Close.rolling(window=n).std()

    data['UpperBand']=data['Moving Average'] + 2 * data['Standard Deviation']
    data['LowerBand']=data['Moving Average'] - 2 * data['Standard Deviation']

    return data
    
```

```python
n=100
bands=BoilingerBand(reliance , n)
```

```python
plt.title('Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.plot(bands.Close,lw=1, label='Close Price')
plt.plot(reliance['UpperBand'],'g',lw=1, label='Upper band')
plt.plot(reliance['Moving Average'],'r',lw=1, label='Middle band')
plt.plot(reliance['LowerBand'],'y', lw=1, label='Lower band')
plt.legend()

plt.show()
```
